```python
#pre-processing 
input["earn_date"] = pd.to_datetime(input['earn_date'])
input = input.sort_values("earn_date")
```

```python
tickers = [
    "AAPL","MSFT","GOOGL","GOOG","AMZN","META","NVDA","AVGO","CRM","ADBE","CSCO",
    "ORCL","TXN","QCOM","AMD","IBM","INTC","AMAT","INTU","MU","ANET","NOW","KLAC",
    "LRCX","SNPS","CDNS","PANW","FTNT","BKNG","NFLX","ADP","PAYX","MSI","HPQ",
    "WDC","ZBRA","GLW","AKAM","ENPH","FSLR","MCHP","MPWR","SWKS","QRVO","TSM",
    "JPM","BAC","WFC","C","GS","MS","BLK","SCHW","CME","ICE","SPGI","MCO","CB",
    "PGR","TRV","AIG","AON","MET","PRU","ALL","AFL","BK","STT","TFC","USB","RF",
    "MTB","HBAN","FITB","CMA","KEY","COF","DFS","PNC","V","MA","PYPL","AXP",
    "BRK.B","TROW","AMP","BEN","IVZ","NTRS","TSLA","HD","MCD","NKE","SBUX","TJX",
    "LOW","TGT","ROST","DG","DLTR","MAR","HLT","EBAY","ETSY","YUM","CMG","CCL",
    "RCL","NCLH","AAP","AZO","ORLY","ULTA","LEN","PHM","DHI","NVR","TOL","POOL",
    "PG","KO","PEP","WMT","COST","MDLZ","MO","PM","KMB","CL","CLX","TAP","STZ",
    "GIS","K","CPB","SJM","KR","WBA","HSY","CHD","BG","ADM","CAT","DE","UPS",
    "FDX","GE","HON","MMM","ITW","EMR","ETN","PH","ROK","GWW","JCI","CARR",
    "OTIS","CSX","NSC","UNP","LUV","DAL","AAL","UAL","MAS","J","IR","NDSN","FAST",
    "SNA","CMI","PCAR","URI","ALLE","AME","AOS","EXPD","XOM","CVX","COP","OXY",
    "EOG","PXD","DVN","MRO","APA","HES","VLO","MPC","PSX","HAL","SLB","BKR",
    "LIN","APD","ECL","SHW","NUE","NEM","FCX","MOS","CF","ALB","MLM","VMC","PKG",
    "IP","WRK","BALL","T","VZ","TMUS","NEE","DUK","SO","SRE","EXC","AEP","XEL",
    "PEG","ED","D","WEC","EIX","PNW","AEE","CMS"
]

test = input[input["ticker"].isin(tickers)]
test
```

```python
##IV filter section - CHANGE TO PRINT MORE CASH
master = test.copy()

##IV_pre RV ratio
master["IV_pre / RV"] = master["IV_pre"] / master["RV30"]
#master = master[master["IV_pre / RV"] >= 17]


#winrate of trade - number of negatives / total
#average win
#average loss
master["percentage_change"] = master["price_change"] / master["price_pre"]
master["win/loss"] = np.where(master["price_change"] < 0, 1, 0) #win/loss column
  
master
```

```python

master_in_sample = master[
    (master["earn_date"] >= "2022-01-01") &
    (master["earn_date"] <= "2024-01-01")
].copy()
master_in_sample = master_in_sample.reset_index(drop=True)

#optimiser
def run_backtest(data, kelly_fraction, starting_amount=100000):
    """
    Runs backtest with sequential compounding and your brokerage logic.
    Returns final account value, total trades, drawdown, equity curve.
    """
    
    account_value = starting_amount
    account_values = [account_value]
    
    for i in range(len(data)):
        
        pct = -1 * data.loc[i, "percentage_change"]   # your notebook formula
        price = data.loc[i, "price_pre"]
        
        position_size = account_value * kelly_fraction
        contracts = (position_size / price).round()
        
        # brokerage both sides
        brokerage_cost = 0.1 * contracts
        trade_pnl = position_size * pct - (brokerage_cost * 2)
        
        account_value += trade_pnl
        account_values.append(account_value)
    
    equity = np.array(account_values)
    running_max = np.maximum.accumulate(equity)
    drawdown = (equity - running_max) / running_max

    return account_value, drawdown.min(), equity


#grid scanned - thresholds are IV/RV, and kelly is kelly criterion
thresholds = np.arange(10, 30, 0.2)
kelly_values = np.linspace(0.02, 0.50, 25)   # 2%â€“50% of account per trade
results = []

#grid search
for T in thresholds:
    
    # filter trades for this IV/RV threshold
    subset = master_in_sample[master_in_sample["IV_pre / RV"] >= T].copy().reset_index(drop=True)
    
    if len(subset) == 0:
        continue
    
    # Compute winrate, avg win, avg loss required for Kelly estimation
    wins = subset[subset["percentage_change"] < 0]     # remember your logic inverts
    losses = subset[subset["percentage_change"] > 0]
    
    if len(wins) == 0 or len(losses) == 0:
        continue
    
    win_rate = len(wins) / len(subset)
    avg_win = (-wins["percentage_change"]).mean()
    avg_loss = (-losses["percentage_change"]).mean()
    
    # theoretical optimal full Kelly fraction
    kelly_full = win_rate - (1 - win_rate) * (avg_win / avg_loss)

    # Instead of relying purely on theoretical Kelly,
    # we BACKTEST many possible Kelly fractions.
    for KF in kelly_values:
        
        final_value, mdd, equity = run_backtest(
            subset,
            kelly_fraction=KF/2,
            starting_amount=100000
        )
        
        results.append([
            T,
            KF,
            kelly_full/2,
            final_value,
            mdd,
            len(subset)
        ])


results_df = pd.DataFrame(
    results,
    columns=["threshold", "half_kelly_fraction", "half_kelly_theoretical", "final_equity", "max_drawdown", "n_trades"]
)

# sort for best result (you can change metric here)
best = results_df.sort_values("final_equity", ascending=False).iloc[0]
print(best)

results_df
```
